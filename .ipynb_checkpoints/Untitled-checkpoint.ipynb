```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataset\\brain_stroke.csv")
df.head()

```

```python
df['gender']=df['gender'].map({'Male':1,'Female':0})
df['ever_married']=df['ever_married'].map({'No':0,'Yes':1})
df['work_type']=df['work_type'].map({'Govt_job':0,'Private':1,'Self-employed':2,'children':3})
df['Residence_type']=df['Residence_type'].map({'Rural':0,'Urban':1})
```

```python
df1=pd.DataFrame({col: df[col].astype('category').cat.codes for col in df}, index=df.index)
```

```python
df1.head(210)
```

```python

X=df1[['gender','age','hypertension','heart_disease','ever_married','work_type','Residence_type','avg_glucose_level','bmi','smoking_status']]
y=df1[['stroke']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_train['stroke'].unique()
```

```python
y_test['stroke'].unique()
```

```python
    rfc = RandomForestClassifier(n_estimators=100)  
    rfc.fit(X_train, y_train)
    prediction=rfc.predict(X_test)
    ranacc=accuracy_score(y_test, prediction)*100
    ranacc
```

```python
    model = AdaBoostClassifier()  
    model.fit(X_train, y_train)
    predi=model.predict(X_test)
    adacc=accuracy_score(y_test, predi)*100
    adacc
```
